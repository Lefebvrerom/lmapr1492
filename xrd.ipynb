# Tâche 7 : Les 3 premiers pics du diffractogramme

Dans cette tâche numéro 7, nous allons déterminer les 3 premiers pics du diffractogramme de notre cristal BaLiP en précisant les indices hkl associés à chaque pic.
Pour toute cette partie, on considère un rayonnement X incident Cu Kα de longueur d’onde λ=1.54060 Å. 
Ces rayons X vont créer un phénomène d'interférences constructives et d'interférences destructives. Les pics du diffractrogramme, correspondent ainsi à des interférences constructives.

```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
```

```python
with MPRester("mxlrouyFDIMHNAE4") as m:
    structure = m.get_structure_by_material_id("mp-10615")
```

```python
xrd = XRDCalculator('CuKa')
pat = xrd.get_pattern(structure)
```

On imprime les pics et leurs intensités.

```python
print(pat)
```

 ## Diffractogramme du BaLiP

Voici le diffractogramme de notre cristal. Celui-ci représente l'angle $2\theta$ en degré en fonction de l'intensité mise à l'échelle (de 0 à 100). Le deuxième graphique est un zoom du premier sur les 3 premiers pics de diffractions.

```python
xrd.show_plot(structure,annotate_peaks=False,two_theta_range=(0,180))
xrd.show_plot(structure,annotate_peaks=True,two_theta_range=(10,35))
```

## Premier pic

```python
print('Le premier pic se trouve à  2*theta = ' + str(pat.x[0]) + '°' )
print('Intensité du premier pic : ' + str(pat.y[0]) + ' %')
print('Indices hkl du premier pic :' + str(pat.hkls[0]))
```

 ## Deuxième pic

```python
print('Le deuxième pic se trouve à  2*theta = ' + str(pat.x[1]) + '°' )
print('Intensité du deuxième pic : ' + str(pat.y[1])+ ' %')
print('Indices hkl du deuxième pic :' + str(pat.hkls[1]))
```
